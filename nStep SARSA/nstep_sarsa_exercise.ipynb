```python
import gym
import numpy as np
import matplotlib.pyplot as plt
from nstep_sarsa_upd2 import SarsaMidterm

# from my_python_file import my_n_step_algorithm

```

```python

env = gym.make('FrozenLake-v1',
               desc=["SFFFFH", "FHFHFF", "FFFHFF", "HFFFGF"],
               is_slippery=False)
# call your algorithm to solve this domain
num_episodes = 1000
instance1 = SarsaMidterm(env,epsilon=0.2, alpha=0.2, gamma = 1, decay_epsilon = True, dyna_alpha= False, toy_game='FrozenLake-v1')
Q, load_policy = instance1.sarsa_nstep(num_episodes,1,1)
instance1.reward_plot()
instance1.step_plot()
```

```python
# Testing part
env = gym.make('FrozenLake-v1',
               desc=["SFFFFH", "FHFHFF", "FFFHFF", "HFFFGF"],
               is_slippery=False,
               render_mode="human")
observation, info = env.reset()
rew=0
num =0
num_episodes = 3
while num<num_episodes:
    action = load_policy[observation]
    observation, reward, terminated, truncated, info = env.step(action)
    rew += reward
    if terminated:
        observation, info = env.reset()
        num += 1
print("total reward:", rew, "Success Rate:", rew/num_episodes*100)
def opt_policy(q):
    policy1 = np.zeros((4,6))
    i=0
    j=0
    for state in q:
        temp = q[state]
        policy1[i,j] = temp
        j+=1
        if (state%6) == 5:
            i+=1
            j = 0
    b=policy1
    c = np.char.array(b,unicode=True)

    for i in range(len(b[:,0])):
        for j in range(len(b[0,:])):
            if b[i,j] == 0:
                c[i,j] = 'L'
            elif b[i,j] == 1:
                c[i,j] ='D'
            elif b[i,j] == 2:
                c[i,j] = 'R'
            else:
                c[i,j] = 'U'
    return c
print(opt_policy(load_policy))
```
